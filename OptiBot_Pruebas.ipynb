# Prototipo de Agente IA (OptiBot)
Este notebook contiene las pruebas iniciales del agente interactuando con el inventario CSV usando LangChain y Cohere, antes de pasarlo a Streamlit.

```python
import pandas as pd
from langchain_cohere import ChatCohere
from dotenv import load_dotenv
import os

# Cargar variables de entorno
load_dotenv()

# Cargar datos
df = pd.read_csv('data/inventario.csv')
df.head()
```

```python
# Inicializar LLM
llm = ChatCohere(temperature=0.2)

# Crear contexto RAG
csv_string = df.to_markdown(index=False)
prompt = f"""Eres OptiBot. Usa este inventario para responder:\n{csv_string}\n\nPregunta: ¿Tienen cajas NAP?"""

respuesta = llm.invoke(prompt)
print(respuesta.content)
```
